```python
import numpy as np
import pandas as pd

stocklist = pd.read_csv('selected_stocks.csv', index_col = 0)
stocklist = stocklist['stock_name']

df = pd.DataFrame(columns= ['stockname', 'avg_profit', 'num_trades'])

for stockname in stocklist :
    df_all = pd.read_csv("stock_data/"+stockname+".csv")
    df_train=df_all[:60000]
    df_train_resampled_60min = df_train[df_train.index%60==0]
    df_train_resampled_60min = df_train_resampled_60min.reset_index(drop=True)   
    df_train_resampled_60min_svwap = df_train_resampled_60min.iloc[:,1]
    df_train_resampled_60min_svwap = df_train_resampled_60min_svwap.to_frame()
    df_train_resampled_60min_svwap = df_train_resampled_60min_svwap.rename(columns={str(list(df_train_resampled_60min_svwap)):0})

    df_new = pd.DataFrame()
#     spread = Top Ask Price - Top Bid Price
    spread = df_train_resampled_60min.iloc[:,4] - df_train_resampled_60min.iloc[:,3]
    avg_spread = np.mean(spread)
#df_new = pd.DataFrame(data,columns=['Deviations'])
    data = np.array([])
    for i in range(262):
        data= np.append(data, df_train_resampled_60min_svwap.iloc[i,0] ) 
    for i in range(262, 1000):
        data= np.append(data, np.mean(df_train_resampled_60min_svwap.iloc[i-262:i+1])) 
    
        df_new = pd.DataFrame(data)
        
    df_deviations = df_train_resampled_60min_svwap.iloc[:,0]-df_new.iloc[:,0] - spread - 0.2
    df_deviations_positive = df_deviations[df_deviations>0]
   # df_deviations_positive.dropna(axis = 0)
    dev_mean = np.mean(df_deviations_positive)
    no_of_trades = df_deviations_positive.size
    
    s=pd.Series([stockname, dev_mean, no_of_trades],index= ['stockname', 'avg_profit', 'num_trades'])
    df=df.append(s,ignore_index=True)
    print(stockname + ":" + str(dev_mean) + ":" + str(no_of_trades))
df.to_csv("profits_2w(14d).csv")
```
